```python
apple_isnull = pandas.isnull(data['Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Apple'])
pumpkin_isnull = pandas.isnull(data['Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Pumpkin'])
pecan_isnull = pandas.isnull(data['Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Pecan'])
ate_pies = apple_isnull & pumpkin_isnull & pecan_isnull
ate_pies.value_counts()
```

```python
def converter(str_value):
    if pandas.isnull(str_value) is True:
        return None
    else:
        str_value = str_value.split(' ')[0]
        str_value = str_value.replace('+', '')
        
        return int(str_value)
data['int_age'] = data['Age'].apply(converter)
data['int_age'].describe()
    
    
```

The age range of participants are fairly even distributed. The lower bound of age is taken to consideration only. 

```python
data['How much total combined money did all members of your HOUSEHOLD earn last year?'].value_counts()
```

```python
def a_integer(str_value):
    if pandas.isnull(str_value) is True:
        return None 
    str_value = str_value.split(' ')[0]
    if str_value == 'Prefer':
        return None
    str_value = str_value.replace('$', '')
    str_value = str_value.replace(',', '')
    return int(str_value)

data['int_income'] = data['How much total combined money did all members of your HOUSEHOLD earn last year?'].apply(a_integer)
data['int_income'].describe()
```

Due to high standard deviation, we can say there is a sharp difference between total income of participants. However the average income is quite high, $75,965. 

```python
travel_u150k = data['int_income'] < 150000
data[travel_u150k]['How far will you travel for Thanksgiving?'].value_counts(normalize = True)
```

```python
travel_u150k = data['int_income'] > 150000
data[travel_u150k]['How far will you travel for Thanksgiving?'].value_counts(normalize = True)
```

Percentage of people with lower income who have to drive for a few hour or less to celebrate Thanksgiving is higher. On the other hand, there is a decent amount of wealthier people who won't travel at all for Thanksgiving. 
Explanation for the first comparision could be students and young adults who do not have high income, work/live in a different city and tend to comeback to there hometown. This is different for the high-income group. They are most likely to settle down at their current place and celebrate Thanksgiving without traveling much. 

```python
data.pivot_table(index = 'Have you ever tried to meet up with hometown friends on Thanksgiving night?', columns = 'Have you ever attended a "Friendsgiving?"', values = 'int_age')
```

Younger people are more willing to attend Friendsgiving or meet up with their hometown friends on that day. 

```python
#Most common dessert (besides pie)
desserts = data.filter(regex = '.*desserts.*')
pandas.value_counts(desserts.values.flatten())
```

The most popular dessert besides pie during Thanksgiving is ice cream. 

```python
#Most common side dishes
side_dish= data.filter(regex = '.*side dishes.*')
pandas.value_counts(side_dish.values.flatten())
```

```python
#Most common pie
pies= data.filter(regex = '.*pie.*')
pandas.value_counts(pies.values.flatten())
```

A common full Thanksgiving meal includes of Turkey, Mashed potatoes, Biscuits, Green bean or Yam casserole and Pumpkin pie. Most of them don't have desserts.

```python
#Regional pattern in dinner menu
regional_main = data[['What is typically the main dish at your Thanksgiving dinner?','What is typically the main dish at your Thanksgiving dinner? - Other (please specify)']].groupby(data['US Region'])
regional_main.describe()
```

```python
regional_side = side_dish.groupby(data['US Region'])
pandas.set_option('display.max_columns', None)
regional_side.describe()
```

The most common side dish in most of the regions is Mashed potatoes. However, in Southern regions such as East South Central, West South Central and South Atlantic, Green beans/green bean casserole, Macaroni and cheese or rolls/biscuits are also very popular. 

```python
regional_pies = pies.groupby(data['US Region'])
pandas.set_option('display.max_columns', None)
regional_pies.describe()
```

Southern regions have more diversed choices of pies while pumpkin remains the most common pie during Thanksgiving.

```python
#Most common dessert besides pie
regional_desserts = desserts.groupby(data['US Region'])
pandas.set_option('display.max_columns', None)
regional_desserts.describe()
```
